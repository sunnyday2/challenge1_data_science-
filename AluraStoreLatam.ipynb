### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



### **Ingreso total por cada tienda**

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.


```python
ventas_tienda1 = tienda1['Precio'].sum()
ventas_tienda2 = tienda2['Precio'].sum()
ventas_tienda3 = tienda3['Precio'].sum()
ventas_tienda4 = tienda4['Precio'].sum()

print('** Ventas **')
print(f'Tienda 1: {ventas_tienda1}')
print(f'Tienda 2: {ventas_tienda2}')
print(f'Tienda 3: {ventas_tienda3}')
print(f'Tienda 4: {ventas_tienda4}')
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Asignar tienda
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

tienda1['Ventas'] = ventas_tienda1
tienda2['Ventas'] = ventas_tienda2
tienda3['Ventas'] = ventas_tienda3
tienda4['Ventas'] = ventas_tienda4

# Función para agrupar por categoría y tienda
def ventas_categoria(productos: pd.DataFrame) -> pd.DataFrame:
    ventas = productos.groupby(['Categoría del Producto', 'Tienda'])[['Producto']].count()
    ventas = ventas.rename(columns={'Producto': 'Cantidad'})
    return ventas

# Obtener ventas por tienda
v1 = ventas_categoria(tienda1)
v2 = ventas_categoria(tienda2)
v3 = ventas_categoria(tienda3)
v4 = ventas_categoria(tienda4)

# Unir todos los DataFrames
ventas_all = pd.concat([tienda1, tienda2, tienda3, tienda4])
ventas_categoria_all = pd.concat([v1, v2, v3, v4])

# Pivotear para tener tiendas como columnas y categorías como filas
ventas_pivot = ventas_categoria_all.reset_index().pivot(index='Categoría del Producto', columns='Tienda', values='Cantidad').fillna(0)

# Ordenar categorías por total descendente
ventas_pivot['Total'] = ventas_pivot.sum(axis=1)
ventas_pivot = ventas_pivot.sort_values(by='Total', ascending=False)
ventas_pivot = ventas_pivot.drop(columns='Total')

# ----------------------------
# Graficar barras horizontales agrupadas
# ----------------------------
plt.figure(figsize=(12, 8))
bar_width = 0.2
y_pos = range(len(ventas_pivot))

for i, tienda in enumerate(ventas_pivot.columns):
    plt.barh(
        [y + i * bar_width for y in y_pos],
        ventas_pivot[tienda],
        height=bar_width,
        label=tienda
    )

# Etiquetas y formato
plt.yticks([y + bar_width * 1.5 for y in y_pos], ventas_pivot.index)
plt.xlabel("Cantidad Vendida")
plt.title("Ventas por Categoría y Tienda")
plt.legend()
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
tienda1_calificacion = round(tienda1['Calificación'].mean(), 2)
tienda2_calificacion = round(tienda2['Calificación'].mean(), 2)
tienda3_calificacion = round(tienda3['Calificación'].mean(), 2)
tienda4_calificacion = round(tienda4['Calificación'].mean(), 2)

print('** Calificaciones promedio por la tienda **')
print(f'Tienda 1: {tienda1_calificacion}')
print(f'Tienda 2: {tienda2_calificacion}')
print(f'Tienda 3: {tienda3_calificacion}')
print(f'Tienda 4: {tienda4_calificacion}')
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
def productos(productos: pd.DataFrame) -> pd.DataFrame:
    max_vendido = int(productos.groupby(['Producto', 'Tienda'])[['Producto']].count().max().iloc[0])
    min_vendido = int(productos.groupby(['Producto', 'Tienda'])[['Producto']].count().min().iloc[0])

    #print(max_vendido)
    #print(min_vendido)

    ventas = productos.groupby(['Producto', 'Tienda'])[['Producto']].count()
    #print(tienda1_ventas)

    ventas.rename(
        columns={
            "Producto": "Cantidad",
        }
        , inplace=True
    )

    tienda_productos = ventas.query('Cantidad in (@max_vendido, @min_vendido)').sort_values(by='Cantidad', ascending=False)
    return tienda_productos

tienda1_productos = productos(tienda1)
tienda2_productos = productos(tienda2)
tienda3_productos = productos(tienda3)
tienda4_productos = productos(tienda4)

print('** Productos más y menos vendidos en cada tienda **')
print(f'\n** Tienda 1:')
print(f'\t\t {tienda1_productos}')

print(f'\n** Tienda 2:')
print(f'\t\t {tienda2_productos}')

print(f'\n** Tienda 3:')
print(f'\t\t {tienda3_productos}')

print(f'\n** Tienda 4:')
print(f'\t\t {tienda4_productos}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
x='Producto'
y='Cantidad'
titulo = 'Productos más y menos vendidos por la Tienda '

# Unir todos los DataFrames
ventas_productos_all = pd.concat([tienda1_productos, tienda2_productos, tienda3_productos, tienda4_productos])

# Pivotear para tener tiendas como columnas y categorías como filas
ventas_productos_pivot = ventas_productos_all.reset_index().pivot(index='Producto', columns='Tienda', values='Cantidad').fillna(0)

# Ordenar categorías por total descendente
ventas_productos_pivot['Total'] = ventas_productos_pivot.sum(axis=1)
ventas_productos_pivot = ventas_productos_pivot.sort_values(by='Total', ascending=False)
ventas_productos_pivot = ventas_productos_pivot.drop(columns='Total')

# ----------------------------
# Graficar barras horizontales agrupadas
# ----------------------------
plt.figure(figsize=(12, 8))
bar_width = 0.2
y_pos = range(len(ventas_productos_pivot))

for i, tienda in enumerate(ventas_productos_pivot.columns):
    plt.barh(
        [y + i * bar_width for y in y_pos],
        ventas_productos_pivot[tienda],
        height=bar_width,
        label=tienda
    )

# Etiquetas y formato
plt.yticks([y + bar_width * 1.5 for y in y_pos], ventas_productos_pivot.index)
plt.xlabel("Cantidad Vendida")
plt.title("Ventas Producto en Tienda")
plt.legend()
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

## **2. Costes de Envío**


El análisis detallado por producto también permitió identificar aquellos que, además de ser muy rentables, presentan costos logísticos bajos. Este tipo de producto debe ser priorizado en futuras campañas comerciales.

```python

# Agrupar por tienda y calcular ingresos y costes de envío
resumen = ventas_all.groupby("Tienda").agg({
    "Precio": "sum",
    "Costo de envío": "sum"
}).reset_index()

# -------------------------------
# Gráfico de líneas – Costes de envío por tienda
# -------------------------------
# Gráfico combinado
fig, ax1 = plt.subplots(figsize=(8, 6))

# Barras para ingresos
ax1.bar(resumen["Tienda"], resumen["Precio"], color='skyblue', label="Ingresos")
ax1.set_ylabel("Ingresos", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Eje secundario para costos de envío
ax2 = ax1.twinx()
ax2.plot(resumen["Tienda"], resumen["Costo de envío"], color='red', marker='o', linewidth=2, label="Costo de envío")
ax2.set_ylabel("Costo de envío", color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Título y leyenda
plt.title("Ingresos vs. Costos de Envío por Tienda")
fig.tight_layout()
plt.show()
```

## **3. Calificaciones de Productos**

El análisis del promedio de calificaciones revela que los productos más vendidos no siempre son los mejor evaluados. Se sugiere revisar de porque el producto **Modelado predictivo** esta mal evaluado por los clientes, dado a su puntuación baja.







```python
# -------------------------------
# Gráfico combinado: Opiniones (Calificación) vs Productos más evaluados
# -------------------------------
# Obtener los productos con más evaluaciones (más registros)
evaluaciones = ventas_all.groupby("Producto").agg({
    "Calificación": ["count", "mean"]
}).reset_index()

evaluaciones.columns = ["Producto", "Cantidad_Evaluaciones", "Promedio_Calificación"]
top_productos = evaluaciones.sort_values("Cantidad_Evaluaciones", ascending=False).head(10)

fig, ax1 = plt.subplots(figsize=(10, 5))

# Barras: Cantidad de evaluaciones
ax1.bar(top_productos["Producto"], top_productos["Cantidad_Evaluaciones"], color="lightgreen", label="Evaluaciones")
ax1.set_ylabel("Cantidad de Evaluaciones", color="green")
ax1.tick_params(axis='y', labelcolor='green')
ax1.set_xticklabels(top_productos["Producto"], rotation=45, ha='right')

# Línea: Promedio de calificación
ax2 = ax1.twinx()
ax2.plot(top_productos["Producto"], top_productos["Promedio_Calificación"], color="red", marker='o', label="Promedio Calificación")
ax2.set_ylabel("Promedio de Calificación", color="red")
ax2.tick_params(axis='y', labelcolor='red')

plt.title("Productos Más Evaluados y Opiniones")
fig.tight_layout()
plt.show()
```

## **4. Vendedores Destacados a Premiar**

Segun el grafico de los mejores vendedores de cada tienda, la vendedora Maria Alonso es quien mas ventas se genero en las Tienda 1 y Tienda 3.

```python
def mejor_vendedor(df: pd.DataFrame) -> pd.DataFrame:
  ventas_por_vendedor = df.groupby(['Tienda', 'Vendedor'])['Precio'].sum().reset_index()
  mejor_por_tienda = ventas_por_vendedor.sort_values('Precio', ascending=False).head(1)
  return mejor_por_tienda

# Unir todos los datos de mejores vendedores en cada tienda
ventas_vendedor = []
for df in [tienda1, tienda2, tienda3, tienda4]:
  ventas_vendedor.append(mejor_vendedor(df))

mejores_vendedores = pd.DataFrame(columns=['Tienda', 'Vendedor', 'Total de Ventas'])
mejores_vendedores[['Tienda', 'Vendedor', 'Total de Ventas']] = pd.concat(ventas_vendedor, ignore_index=True)
mejores_vendedores.sort_values(by='Total de Ventas', inplace=True, ascending=False)
mejores_vendedores.reset_index(drop=True, inplace=True)
print(mejores_vendedores)


# Gráfico de torta
plt.figure(figsize=(8, 6))
plt.pie(mejores_vendedores['Total de Ventas'], labels=mejores_vendedores['Tienda'] + ':' + mejores_vendedores['Vendedor'], autopct='%1.1f%%', startangle=140)
plt.title('Top Mejores Vendedores de cada Tienda)')
plt.axis('equal')  # Asegura que el gráfico sea un círculo
plt.show()
```
